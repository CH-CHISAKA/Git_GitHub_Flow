```python
import requests
from bs4 import BeautifulSoup

def print_grid_from_doc(url):
    html = requests.get(url).text
    soup = BeautifulSoup(html, "html.parser")

    rows = soup.find_all("tr")[1:]
    points = []

    for row in rows:
        cells = [cell.get_text(strip=True) for cell in row.find_all("td")]
        if len(cells) == 3:
            x = int(cells[0])
            char = cells[1]
            y = int(cells[2])
            points.append((x, y, char))

    max_x = max(x for x, y, char in points)
    max_y = max(y for x, y, char in points)

    grid = [[" " for _ in range(max_x + 1)] for _ in range(max_y + 1)]

    for x, y, char in points:
        grid[y][x] = char

    for row in grid:
        print("".join(row))


print_grid_from_doc("https://docs.google.com/document/d/e/2PACX-1vSvM5gDlNvt7npYHhp_XfsJvuntUhq184By5xO_pA4b_gCWeXb6dM6ZxwN8rE6S4ghUsCj2VKR21oEP/pub")
```
